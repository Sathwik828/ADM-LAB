```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
file_path = "employee_performance.csv"  # Ensure this file exists in your directory
df = pd.read_csv('/content/employee_performance.csv')
```

```python
# Display statistical summaries
print("Statistical Summary:")
print(df.describe())
```

```python
# Data Transformation and Aggregation
# Encoding categorical column
df['Department'] = df['Department'].astype('category').cat.codes
```

```python
# Creating a new derived column (e.g., Salary per Year of Experience)
df['Salary_per_Year_Experience'] = df['Monthly_Salary'] * 12 / df['Years_of_Experience'].replace(0, 1)
```

```python
# Grouping and Aggregation: Average salary per department
salary_per_department = df.groupby('Department')['Monthly_Salary'].mean()
print("\nAverage Monthly Salary per Department:")
print(salary_per_department)
```

```python
# Visualization
sns.set(style="whitegrid")
```

```python
# Histogram of Monthly Salaries
plt.figure(figsize=(8, 5))
sns.histplot(df['Monthly_Salary'], bins=20, kde=True, color='blue')
plt.title('Histogram of Monthly Salaries')
plt.xlabel('Monthly Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter Plot: Years of Experience vs. Performance Score
plt.figure(figsize=(8, 5))
# Assign the scatterplot to a variable
scatter = sns.scatterplot(x=df['Years_of_Experience'],
                          y=df['Performance_Score'],
                          hue=df['Monthly_Salary'],
                          palette='coolwarm')
plt.title('Years of Experience vs. Performance Score')
plt.xlabel('Years of Experience')
plt.ylabel('Performance Score')
# Use the scatterplot object to create the colorbar
plt.colorbar(scatter.collections[0], label='Monthly Salary') # Pass the mappable object
plt.show()
```

```python
# Heatmap of numerical features
plt.figure(figsize=(8, 5))
sns.heatmap(df[['Monthly_Salary', 'Years_of_Experience', 'Performance_Score']].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap of Numerical Features')
plt.show()
```
